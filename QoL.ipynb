```r
library(repr)
library(ggplot2)
library(ggrepel)
```

```r
numbeo = read.csv("qol-numbeo.csv")
head(numbeo)
```

```r
options(repr.plot.width = 28, repr.plot.height = 16)
ggplot(numbeo, aes(x=Cost.of.Living.Index, y=Quality.of.Life.Index, label=City)) + geom_point() + geom_text_repel()
```

```r
mercer = read.csv("qol-mercer.csv")
head(mercer)
```

```r
options(repr.plot.width = 28, repr.plot.height = 16)
ggplot(mercer, aes(x=-Cost.Rank, y=-Quality.Rank, label=City)) + geom_point() + geom_text_repel()
```

```r
options(repr.plot.width = 28, repr.plot.height = 16)
ggplot(
    mercer, 
    aes(
        x=1/Cost.Rank,
        y=1/Quality.Rank,
        label=City
    )) + 
    geom_point() + 
    geom_text_repel() + 
    scale_x_continuous(trans='log2') +
    scale_y_continuous(trans='log2')
```
